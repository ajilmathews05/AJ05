# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
from enum import auto
WCS=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(df)
  WCS.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),WCS)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42,n_init='auto')
y_pred=kmeans.fit_predict(df)
```

```python
y_pred
```

```python
plt.scatter(df.iloc[y_pred == 0, 0], df.iloc[y_pred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(df.iloc[y_pred == 1, 0], df.iloc[y_pred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(df.iloc[y_pred == 2, 0], df.iloc[y_pred == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
