```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Students_Performance.csv')
df
```

**An Institution wishes to find out their student’s ability in
mathematics, reading and writing skills. The Institution
wants to do an exploratory study to check the following
information.**

**A. Find out how many males and females participated in
the test.**


```python
df['gender'].value_counts()
```

**B. What do you think about the students' parental level of
education? ***


```python
df['parental level of education'].value_counts()
```

**C.Who scores the most on average for math, reading and
writing based on**

**i. Gender**

```python
df.dropna(subset=['math score','reading score','writing score'],inplace=True)
```

```python
gender_scores= df.groupby('gender')[['math score','reading score','writing score']].mean()
gender_scores
```

**Who scores the most on average for math, reading and writing based on**

**ii. Test preparation course**


```python
test_preparation_course= df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
test_preparation_course
```

**D. What do you think about the scoring variation for
math, reading and writing based on**

**i. Gender**

```python
gender_scores= df.groupby('gender')[['math score','reading score','writing score']].var()
gender_scores
```

**D. What do you think about the scoring variation for math, reading and writing based on**

**ii. test preparation course**

```python
test_preparation_course= df.groupby('test preparation course')[['math score','reading score','writing score']].var()
test_preparation_course
```

**E.The management needs your help to give bonus points
to the top 25% of students based on their math score,
so how will you help the management to achieve this.**

```python
top=df[df['math score'] >= df['math score'].quantile(0.75)]
print(top)
```
